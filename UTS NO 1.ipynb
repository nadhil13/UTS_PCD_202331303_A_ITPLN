## IMPORT LIBRARY

```python
## Pemisahan Channel Warna
# 202331303_Muhammad Nadhil Arsy Al-Wafi
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## BACA GAMBAR

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
# Membaca gambar
img1 = cv2.imread('nama.jpg')
img2 = cv2.imread('info.jpg')

# Mengonversi warna BGR ke RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Membuat subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Menampilkan gambar pertama
axs[0].imshow(img1_rgb)
axs[0].axis('off')

# Menampilkan gambar kedua
axs[1].imshow(img2_rgb)
axs[1].axis('off')

plt.show()
```

## PISAHKAN CHANNEL B, G, R

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
b, g, r = cv2.split(img1)
```

# Penerapan threshold untuk mendeteksi warna biru, merah, dan hijau
# Nilai threshold ini perlu disesuaikan berdasarkan analisis histogram

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi

# Threshold untuk warna biru
_, blue_mask = cv2.threshold(b, 150, 255, cv2.THRESH_BINARY)

# Threshold untuk warna hijau
_, green_mask = cv2.threshold(g, 160, 255, cv2.THRESH_BINARY)

# Threshold untuk warna merah
_, red_mask = cv2.threshold(r, 170, 255, cv2.THRESH_BINARY)
```

## VISUALISASI HASIL DETEKSI

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi

# Tampilkan hasil
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('CITRA KONTRAS')

plt.subplot(2, 2, 2)
plt.imshow(blue_mask, cmap='gray')
plt.title('BIRU')

plt.subplot(2, 2, 3)
plt.imshow(red_mask, cmap='gray')
plt.title('MERAH')

plt.subplot(2, 2, 4)
plt.imshow(green_mask, cmap='gray')
plt.title('HIJAU')

plt.tight_layout()
plt.show()
```

## PEMBUATAN HISTOGRAM UNTUK ANALISIS

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi

# MENAMPILKAN HISTOGRAM UNTUK DETEKSI WARNA
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Cara 1 - Channel Blue
axs[0, 0].imshow(b, cmap='gray')
axs[0, 0].set_title('Channel Blue')
axs[0, 1].hist(b.ravel(), 256, [0, 256], color='blue')
axs[0, 1].set_title('Histogram Blue')

# Cara 2 - Channel Green
axs[1, 0].imshow(g, cmap='gray')
axs[1, 0].set_title('Channel Green')
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
axs[1, 1].plot(hist_g, color='green')
axs[1, 1].set_title('Histogram Green')

# Cara 3 - Channel Red
axs[2, 0].imshow(r, cmap='gray')
axs[2, 0].set_title('Channel Red')
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
axs[2, 1].plot(hist_r, color='red')
axs[2, 1].set_title('Histogram Red')

# Menampilkan
plt.tight_layout()
plt.show()
```
